H1: There is a statistically significant difference between converion rates of the average bidding and maximum bidding.

## A/B Testing (Independent two sample t-test)

Independent two sample t-test: It is used to compare the means of two different groups. Independent two sample t-test aims to prove the observed difference between means is statistically and significantly different. The analysis should start with (1) setting up the hypothesis and followed by (2) checking the assumptions and (3) interpretation of the p-value.

### Hypothesis of the A/B Test

H0: µ1 = µ2 <br />
H1: µ1 != µ2 

## Assumptions

- 1. Normality Assumption: Distributions of the data groups should be normally distributed. Outlier analsis is important to get normally distributed sample (sahpiro-wilks test). Shapiro-wilks test checks whether there is a statistical diffirence between standard normal distribution and distribution of the variable.

    H0: Variable is normally distributed.<br />
    H1: Variable is not normally distributed.
    
- 2. Variance Homogeneity: Levene test is used to cehck the variance homogeneity, it checks the homogeneity of the considered two samples, in other words it checks the similarity of the variances of the considered two samples.

    H0: Variances are homogenous.<br />
    H1: Variances are not homogenous.

If these assumptions are fulfilled, independent two sample t-test (parametric test) should be implemented. Otherwise, mannwihtneyu test (non-parametric test) should be implemented.

If normality assumption is not fulfilled, samples should be analysed with mannwhitneyu test. If variance homogeneity is not observed, samples are analysed with independent two sample t-test by changing the parameter (equal_var = False).   


```python
# Checking normality assumption
test_stat, pvalue = shapiro(cont_df["Purchase"])
print("Test Stat = %.4f, p-value = %4f" % (test_stat, pvalue))
```

```python
test_stat, pvalue = shapiro(test_df["Purchase"])
print("Test Stat = %.4f, p-value = %4f" % (test_stat, pvalue))
```

**Findings show that H0 can not be rejected (p>0.05) for both of the variables, which means that both variables are normally distributed.**

```python
# Checking variance homogeneity 
test_stat, pvalue = stats.levene(cont_df["Purchase"], test_df["Purchase"])
print("Test Stat = %.4f, p-value = %.4f" % (test_stat, pvalue))
```

**Finding show that H0 can not be rejected, so the variables have equal variances. Since two assumptions are fulfilled, independent two sample t-test is used to compare the means of two independent samples.** 

```python
test_stat, pvalue = stats.ttest_ind(cont_df["Purchase"], test_df["Purchase"],
                             equal_var=True)
print("Test Stat = %.4f, p-value = %.4f" % (test_stat, pvalue))
```

P-value is greater that 0.3493, so the observed two samples have an identical average values. This means that there is no statistically significant differnce between conversion rates of maximum bidding and average bidding.
